# Resumen Logistic Regression

La regresión logística es un algoritmo que permite la clasificación de un conjunto de datos en 2 o mas grupos.

Observando la imagen, supongamos que tenemos un set de datos conformado por 2 columnas X1 y X2, claramente diferenciados en 2 grupos, y que a su vez le hemos asignado una característica a cada par de puntos (X1,X2), por ejemplo a todos los blancos 0 y a todos los negros 1 siendo esta la variable "y". 

En la imagen de la izquierda vemos X1 y X2, puntos blancos en un grupo y puntos negros en otro.

A su vez observamos, por ejemplo, que el punto "p6", tiene como valores de X1=5 y de X2=1.5, su característica y=0.
Podemos observar que para el grupo 1 (negros) los valores de X1 y X2 son predomiantemente mas altos que los del grupo 0 (blancos)

Al calcular con la regresión logística "y" tomará valores por encima de 0.5 para cada punto negro y por debajo de 0.5 para cada punto blanco.

La regresión logística devolverá una matriz de $\theta_1$ y $\theta_2$ que son los factores que multiplicando a las X1 y X2 devuelven un valor entre 0 y 1 que nos permitirá definir si para ese par X1,X2 es un punto negro(1) o blanco(0). A éste proceso lo llammos predicción.

De ésta manera podemos definir que para todo valor menor de 0.5 asignaremos 0 a "y" y para todo valor mayor a 1 asignaremos 1 a "y", como se observa a la derecha de la imagen.





## Types of Logistic Regression

There Are Three Types:

a- Binomial

b- Ordinal

c- Multinomial

### a Binomial

Binomial Logistic regression deals with those problems with target variables having only two possible values, 0 or 1.

Which can Signify Yes/No, True /False, Dead/Alive, and other categorical values.

### b Ordinal

Ordinal Logistic Regression Deals with those problems whose target variables can have 3 or more than 3 values, unordered in nature. Those values don’t have any quantitative significance

For Example Type 1 House, Type 2 House, Type 3 House, etc

### c Multinomial

Multinomial Logistic regression, just Ordinal Logistic Regression, deals with Problems having target values to be more than or equal to 3. The main difference lies that unlike Ordinal, those values are well ordered. The values Hold Quantitative Significance
